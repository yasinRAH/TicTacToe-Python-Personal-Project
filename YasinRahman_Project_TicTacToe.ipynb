```python
from IPython.display import clear_output

def game_board(board):
    clear_output()
    
    print(board[7] + "|" + board[8] + "|" + board[9])
    print("- - -") 
    print(board[4] + "|" + board[5] + "|" + board[6])
    print("- - -")
    print(board[1] + "|" + board[2] + "|" + board[3])
```

```python
board = [None, 
         " ", " ", " ", 
         " ", " ", " ", 
         " ", " ", " "]
#resets board
def board_reset():
    board[1::] = [" "] * (len(board) - 1)
    return board

#player classes
def player_x():
    while True:   
        try:
            x_input = int(input("X, choose a cell corresponding to the numpad "))
        except ValueError:
            print("Invalid position! The input needs to be a number")
            continue
        if 1 <= x_input <= 9 and board[x_input] == " ":
            board[x_input] = "X"
            break  
        elif 1 > x_input > 9:
            print("Invalid position! The number does not correspond with a cell")
        else:
            print("Invalid position! There is already a piece there")
    
    return game_board(board)

def player_o():
    while True:   
        try:
            o_input = int(input("O, choose a cell corresponding to the numpad "))
        except ValueError:
            print("Invalid position! The input needs to be a number")
            continue
        if 1 <= o_input <= 9 and board[o_input] == " ":
            board[o_input] = "O"
            break  
        elif 1 > o_input > 9:
            print("Invalid position! The number does not correspond with a cell")
        else:
            print("Invalid position! There is already a piece there")
            
    return game_board(board)
```

```python
#create checks for win 
#seperate board list into chunks of 3
def make_original_matrix(board):  
    original_board_matrix = [board[i:i+3] for i in range(1, len(board), 3)]
    return original_board_matrix

#creates matrix to check columns/verticals
def make_verticals(board_matrix):
    col_1, col_2, col_3 = [], [], []
    for i in board_matrix:
        col_1.append(i[0])
        col_2.append(i[1])
        col_3.append(i[2])

    verticals_matrix = [col_1, col_2, col_3]
    return verticals_matrix

#takes diagonal 1 from board
def diagonal_1():
    diag_1 = []
    count = 0
    for i in make_original_matrix(board):
        diag_1.append(i[count])
        count += 1
    return diag_1 
    
#takes diagonal 2 from board
def diagonal_2():
    diag_2 = []
    count = 2
    for i in make_original_matrix(board):
        diag_2.append(i[count])
        count -= 1
    return diag_2

#checks if all values in diagonals match
def diagonal_repeat_check(diagonal):
    compare_string = "".join(diagonal)
    if compare_string == "XXX":
        game_ended = True
        return "X wins!"
    elif compare_string == "OOO":
        game_ended = True
        return "O wins!"
    else:
        return " "

#looks at matrix and checks for repeats 
def matrix_repeat_check(matrix):
    for i in matrix:
        compare_string = "".join(i)
        if compare_string == "XXX":
            game_ended = True
            return "X wins!"
        elif compare_string == "OOO":
            game_ended = True
            return "O wins!"
    return " "


#checks all win conditions
def game_end_check():
    if matrix_repeat_check(make_original_matrix(board)) != " ":
        return matrix_repeat_check(make_original_matrix(board))
    elif matrix_repeat_check(make_verticals(make_original_matrix(board))) != " ":
        return matrix_repeat_check(make_verticals(make_original_matrix(board)))
    elif diagonal_repeat_check(diagonal_1()) != " ":
        return diagonal_repeat_check(diagonal_1())
    elif diagonal_repeat_check(diagonal_2()) != " ":
        return diagonal_repeat_check(diagonal_2()) 
    elif " " not in board:
        return "Tie!"
    else:
        return False

    
```

```python
def game():
    def game_start():
        
        while True:
            player_x()
            if game_end_check() is not False:
                break
            player_o()
            if game_end_check() is not False:
                break
            
        print(game_end_check())
    
    def board_setup():
        board_reset()
        game_board(board)
        game_start()
            
    while True:
        ask_to_play = input("Play, YES or NO? ")
        if ask_to_play.upper() == "YES" or ask_to_play.upper() == "Y": 
            board_setup()
            while True:
                repeat = input("Play again, YES or NO? ")
                if repeat.upper() == "YES" or repeat.upper() == "Y":  
                    board_setup()
                elif repeat.upper() == "NO" or repeat.upper() == "N":
                    print("Bye!")
                    return
                else:
                    print("Sorry, I didn't catch that ")
                    continue
        elif ask_to_play.upper() == "NO" or ask_to_play.upper() == "N":
            print("Bye!")
            return 
        else:
            print("Sorry, I didn't catch that ")
            continue    
game()
```
